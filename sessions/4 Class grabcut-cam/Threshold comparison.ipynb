```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.append('../../')
import utils.path as path
```

```python
gccam_path = path.goback_from_current_dir(0) + r'threshold_0'
cam_path = path.goback_from_current_dir(1) + r'3 Class cam\threshold_0'
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
plt.rcParams["figure.figsize"] = (20, 12)
```

```python
mIoU = []
mAcc = []

for i in range(1, 10):
    df = pd.read_csv(gccam_path + f'{i}\IoU.csv')
    mIoU.append(df['IoU'].mean())
    mAcc.append(df['Accuracy'].mean())
    
df = pd.DataFrame(list(zip(mIoU, mAcc)),
           columns =['mIoU', 'mAccuracy'])
```

```python
fig, ax = plt.subplots()

ind = np.arange(len(df))
width = 0.3   
ax.bar(ind, df['mIoU'], width, label = "mIoU")
ax.bar(ind + width, df['mAccuracy'], width, label = "mAccuracy")
ax.set_xlabel('Thresholds', fontsize = 12)
ax.set_title(f'gccam - mIoU and mAccuracy by threshold', fontsize = 18)
ax.legend()
plt.xticks(ind + width / 2, ((df.index + 1) / 10))

plt.show()
fig.savefig(path.goback_from_current_dir(0) + "gccam - mIoU and mAccuracy by threshold.jpg")
```
