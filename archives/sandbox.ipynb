```python
from PIL import Image
from torchvision import transforms
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
preprocess = transforms.Compose([
   transforms.ToTensor()])
```

```python
img = Image.open(r'bottles.png')
img = preprocess(img)
```

```python
print(img.shape)
```

```python
img = img.numpy()
```

```python
print(np.unique(img * 255))
```

```python
print(np.max(img))
```

```python
def pil_to_cv2(img):
    cv2_image = np.moveaxis(img, 0, -1)
    return (cv2_image * 255).astype('uint8')
```

```python
img = pil_to_cv2(img)
```

```python
img[:, :, 1] = 0
```

```python
print(img.shape)
```
