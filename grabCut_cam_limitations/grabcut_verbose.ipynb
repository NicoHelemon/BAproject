```python
import numpy as np
import cv2
import os
import json
import utils
from matplotlib import pyplot as plt
```

```python
target_directory = r'output\session_4\grabcut_verbose\\'
```

```python
image = cv2.imread(r'images\original_images\set_1\cat.jpg')
```

```python
image = utils.resize_target_height(image, 600)
image = image[:, :800]
```

```python
utils.pltShow(image)

cv2.imwrite(target_directory + f'cat.jpg', image)
```

```python
bbox = [150, 90, 435, 505]
```

```python
image_bbox =  utils.drawBbox(image, bbox, color = (255, 255, 255), thickness = 2)
```

```python
utils.pltShow(image_bbox)

cv2.imwrite(target_directory + f'BBOX_cat.jpg', image_bbox)
```

```python
inMask = utils.bboxToMask(image, bbox, cv2.GC_BGD)
outMask, binMask = utils.grabCut(image, inMask, mode = 'MASK')
output = cv2.bitwise_and(image, image, mask = binMask)

inMask = utils.maskToGrayImage(inMask)
outMask = utils.maskToGrayImage(outMask)

utils.pltShow(inMask, mode = 'gray')
utils.pltShow(outMask, mode = 'gray')
utils.pltShow(binMask, mode = 'gray')
utils.pltShow(output)

cv2.imwrite(target_directory + f'DEF_IN_MASK_cat.jpg', inMask)
cv2.imwrite(target_directory + f'DEF_OUT_MASK_cat.jpg', outMask)
cv2.imwrite(target_directory + f'DEF_BIN_MASK_cat.jpg', binMask)
cv2.imwrite(target_directory + f'DEF_OUTPUT_cat.jpg', output)
```

```python
inMask = utils.bboxToMask(image, bbox, cv2.GC_PR_BGD)
outMask, binMask = utils.grabCut(image, inMask, mode = 'MASK')
output = cv2.bitwise_and(image, image, mask = binMask)

inMask = utils.maskToGrayImage(inMask)
outMask = utils.maskToGrayImage(outMask)

utils.pltShow(inMask, mode = 'gray')
utils.pltShow(outMask, mode = 'gray')
utils.pltShow(binMask, mode = 'gray')
utils.pltShow(output)

cv2.imwrite(target_directory + f'PROB_IN_MASK_cat.jpg', inMask)
cv2.imwrite(target_directory + f'PROB_OUT_MASK_cat.jpg', outMask)
cv2.imwrite(target_directory + f'PROB_BIN_MASK_cat.jpg', binMask)
cv2.imwrite(target_directory + f'PROB_OUTPUT_cat.jpg', output)
```
